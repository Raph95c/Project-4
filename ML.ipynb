```python
import pandas as pd
import hvplot.pandas
from pathlib import Path
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Read the CSV file into a Pandas DataFrame
# Set the index using the Ticker column
ml_df= pd.read_csv("/Users/raphoun/Documents/GitHub/Project3_group/Project-4/Data/tenyear.csv" ,index_col=[0])


# Review the DataFrame
ml_df
```

```python
df=ml_df.drop(['artists','explicit', 'mode','release_date','name','year'], axis=1, inplace=False)
df
```

```python
df.set_index(['id'])
```

```python
 # Seperate the features, X,  from the target variable, y
y = df['duration_ms']
X = df.drop(columns='duration_ms')
```

```python
# Preview the features data
X.head()
```

```python
y[:5]
```

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Instantiate a StandardScaler instance
scaler = StandardScaler()

# Fit the training data to the standard scaler
X_scaler = scaler.fit(X_train)

# Transform the training data using the scaler
X_train_scaled = X_scaler.transform(X_train)

# Transform the testing data using the scaler
X_test_scaled = X_scaler.transform(X_test)
```

```python
 # Import the KNeighborsClassifier module from sklearn
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the KNeighborsClassifier model with n_neighbors = 3 
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model using the training data
knn.fit(X_train_scaled, y_train)
```

```python
 # Create predictions using the testing data
y_pred = knn.predict(X_test_scaled)
```

```python
# Print the classification report comparing the testing data to the model predictions
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
