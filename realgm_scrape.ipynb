```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
url = "https://basketball.realgm.com/nba/players"
browser = Browser("chrome")
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
table = soup.find("table", class_="tablesaw tablesaw-swipe tablesaw-sortable")
thead = table.find("thead")
tbody = table.find("tbody")
```

```python
row_headers = []
for x in thead.find_all("tr"):
    for y in x.find_all("th", class_="tablesaw-sortable-head"):
        row_headers.append(y.text)

print(row_headers)
```

```python
players = [] 
player_data = []

for x in tbody.find_all("tr"):
    if len(player_data) > 1:
        players.append(player_data)
        player_data = []
    for y in x.find_all("td"):
        player_data.append(y.text)
```

```python
players_df = pd.DataFrame(players, columns=row_headers)
players_df.head(5)
```

```python
players_df2 = players_df.convert_dtypes()
players_df2.dtypes
```

```python
players_df2 = players_df2.astype({"YOS": "int"
                                  ,"WT": "float"
                                  ,"#" : "float"})
```

```python
five_yr_baller = players_df2[players_df2['YOS'] >= 5]
print(five_yr_baller.to_string())
```
